1. Start with a simple free-stream at an arbitrary angle of attack which is specified by the angle of attack Î±. Write out the mathematical equation for its potential and streamfunction in cartesian form without complex analysis. Also write out the expression for the complex potential.

### Velocity Field
Assuming angle of attack a=30 degrees

```python
a=np.pi/6
x, y = np.mgrid[-5.0:5.0:20j, -5:5:20j]
u=np.cos(a)
v=np.sin(a)
plt.quiver(x, y, u, v);
```

### Potential

```python
def potential(x,y):
    return x*np.cos(a)+y*np.sin(a)
plt.contour(x, y, potential(x,y));
plt.axis('equal');
```

### Streamfunction

```python
def streamfunction(x,y):
    return -x*np.sin(a)+y*np.cos(a)
plt.contour(x, y, streamfunction(x,y));
plt.axis('equal');
```

### Velocity field

```python
x, y = np.mgrid[-5.0:5.0:20j, -5:5:20j]
u1=1+(x+0.5)/2*np.pi*(np.sqrt((x+0.5)**2+y**2))-(x-0.5/2)*np.pi*np.sqrt((x-0.5)**2+y**2)
v1=y/2*np.pi*(np.sqrt((x+0.5)**2+y**2))-(y/2)*np.pi*np.sqrt((x-0.5)**2+y**2)
r=(x**2+y**2)/2
u1[r<0.01]=0
v1[r<0.01]=0
plt.quiver(x,y,u1,v1)
plt.axis('equal')
```

### Streamlines and potential lines

```python
x, y = np.mgrid[-2.0:2.0:50j, -2:2:50j]
def potential(x,y):
    return x+np.log(np.sqrt((x+0.5)**2+y**2))/(2*np.pi)-np.log(np.sqrt((x-0.5)**2+y**2))/(2*np.pi)
plt.contour(x, y, potential(x,y));
plt.axis('equal');
```
