```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Consider the flow past a circular cylinder of radius 1, with the free stream from the left side.

1) Compute the coefficient of pressure, Cp along the surface of the cylinder as a function of θ from the x-axis. Plot this coefficient of pressure. Also plot the surface velocity magnitude distribution as a function of θ.

ComplexPotential = U(z+R^2/z)

Since U=1, R=1

ComplexPotential = z+1/z

Complex velocity = d(streamfunction)/dz

Complex velocity =1-1/z^2

Since z = r*e^i*theta, and theta = tan^-1(y/x)

Complex velocity = 1-(e^-i*2*theta)/r^2

e^-2i*theta = cos(2*theta)-i*sin(2*theta)

Complex velocity = u-iv

u = 1-cos(2*theta)/r^2

v= -sin(2*theta)/r^2

Net velocity magnitute = 1+1/r^4-2*cos^2(2*theta)/r^2

Coefficient of pressure(Cp) = 1-(net velocity magnitude/strength of streamfunction)^2=2*cos^2(2*theta)/r^2 - 1/r^4

For surface of cylinder, Cp = 2*cos^2(2*theta)-1

Surface velocity magnitude = 2-2*cos^2(2*theta)

```python
theta=np.mgrid[-360:360:720j]
x, y = [np.cos(np.deg2rad(theta)), np.sin(np.deg2rad(theta))]
z=x+1j*y 
complex_velocity=1-1/(z**2)
V=abs(complex_velocity)
u_square=V**2
c_p=1-V**2
plt.plot(theta,c_p)
plt.xticks(list(np.mgrid[-360:360:9j]))
plt.xlabel('θ (degrees)')
plt.ylabel('Cp')
plt.grid()
plt.title('Cp at Surface v/s θ')
```

```python
plt.plot(theta,np.sqrt(u_square))
plt.xlabel('θ (degrees)')
plt.ylabel('Velocity')
plt.xticks(list(np.mgrid[-360:360:9j]))
plt.grid()
plt.title('Surface Velocity magnitude v/s θ')
```

Pressure=1-(1.225/2)*v^2

Complexvelocity = 1 - (1/z^2) - (Ti/2*pi*z)

pressure = 1 - (1.225/2)*abs(complexvelocity)*2

```python
T=2
complex_velocity=1-(1/z**2)-(T*1j/(2*np.pi*z))
pressure=1-((1.225/2)*(abs(complex_velocity))**2)
plt.contour(x, y, pressure,int(25));
plt.axis('equal');
plt.colorbar()
plt.grid()
plt.title('Pressure using Complex Velocity for T=2')
```
