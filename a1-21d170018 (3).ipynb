```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

# Question 1

Consider the velocity field given by:

u=((x^2−y^2−1)(x^2−y^2)+4x^2*y^2)/r^4
v=−2xy/r^4
r=√(x^2+y^2)

Further, set u,v be set to zero when r<1. Plot the velocity field using lines with arrows. Plot the streamlines. also plot path lines and streaklines for 10 points that start from the line x = -2.
What does this flow look like to you?

## Solution

```python
def vel_u(x,y):
    flag = 1
    r = math.sqrt(x*x + y*y)
    if r<1:
        flag = 0
    if flag == 1:
        u =((x*x - y*y - 1)*(x*y - y*y)+4*x*x*y*y)/(x*x+y*y)**2
    else:
        u=0
    return u
def vel_v(x,y):
    flag = 1
    r = math.sqrt(x*x+y*y)
    if r<1:
        flag = 0
    if flag == 1:
        v=-2*x*y/(x*x+y*y)**2
    else:
        v = 0
    return v
```

```python
x = np.mgrid[-2:2:10j]
y = np.mgrid[-2:2:10j]
for i in range(10):
    for j in range(10):
        u = vel_u(x[i],y[j])
        v = vel_v(x[i],y[j]) 
        plt.quiver(x[i],y[j],u,v)
```

Pathlines/Streamlines

```python
def vel(r,t):
    u=vel_u(r[0],r[1])
    v=vel_v(r[0],r[1])
    return np.array([u,v])
times = np.linspace(0, 10, 1000)
for i in range(10):
    ic = np.array([-2,i*0.5])
    sol = odeint(vel, ic, times)
    sx, sy = sol.T
    plt.plot(sx, sy)
plt.axis('equal')
```

Streaklines

```python
def vel_u(x,y):
    flag = 1
    r = math.sqrt(x*x + y*y)
    if r<1:
        flag = 0
    if flag == 1:
        u =((x*x - y*y - 1)*(x*y - y*y)+4*x*x*y*y)/(x*x+y*y)**2
    else:
        u=0
    return u
def vel_v(x,y):
    flag = 1
    r = math.sqrt(x*x+y*y)
    if r<1:
        flag = 0
    if flag == 1:
        v=-2*x*y/(x*x+y*y)**2
    else:
        v = 0
    return v
def vel(r,t):
    u=vel_u(r[0],r[1])
    v=vel_v(r[0],r[1])
    return np.array([u,v])
ic = np.array([-2.0, 0.0])
dt = 0.1
n = 100
T = n*dt
sx, sy = [], []
for j in range(10):
    ic=np.array([-2.0,j])
    for i in range(n):
        start = i*dt
        res = odeint(vel, ic, [start, T])
        sx.append(res[1,0])
        sy.append(res[1,1])
plt.plot(sx, sy,linewidth=2.0)
plt.grid()
```

As seen from the streaklines, the particles are return back to the line x=-2.
